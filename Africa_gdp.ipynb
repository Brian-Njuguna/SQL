```python
%%sql
SELECT 
    wb.country_name,
    r.region_name, 
    r.sub_region_name
FROM wb_africa wb
LEFT JOIN  regions r
    ON wb.country_name = r.country_name
limit 10;
```

```python
%%sql
SELECT DISTINCT
    Country_name,
    RANK()OVER( PARTITION BY  year ORDER BY gdp_per_capita DESC) AS rankings,
    Year,
    gdp_per_capita,
    gdp_per_capita_billion
FROM wb_africa
WHERE year IN ('2023','2024')
ORDER BY rankings
limit 10;
```

```python
%%sql
SELECT DISTINCT
    wb.Country_name,
    wb.Year,
    gdp_per_capita,
    r.Least_Developed_Countries,
    gdp_per_capita_billion,
    RANK() OVER(PARTITION BY wb.year ORDER BY wb.gdp_per_capita DESC) As gdp_rank
FROM wb_africa wb
LEFT JOIN regions r
     ON r.country_name = wb.country_name
WHERE r.least_developed_countries != ''
AND    wb.year = '2024'
ORDER BY gdp_rank ASC;
```

```python
%%sql
SELECT 
    Country_name,
    CONCAT(ROUND(gdp_per_capita / 1000000000, 2), 'B') AS gdp_in_billions
FROM wb_africa
limit 10;
```

```python
%%sql
SELECT
    Country_name,
    MAX(CASE WHEN year = '2020' THEN gdp_per_capita_billion END) AS GDP_2020,
    MAX(CASE WHEN year = '2021' THEN gdp_per_capita_billion END) AS GDP_2021,
    MAX(CASE WHEN year = '2022' THEN gdp_per_capita_billion END) AS GDP_2022,
    MAX(CASE WHEN year = '2023' THEN gdp_per_capita_billion END) AS GDP_2023,
    MAX(CASE WHEN year = '2024' THEN gdp_per_capita_billion END) AS GDP_2024
FROM wb_africa
GROUP BY Country_name
limit 10;
```

```python
%%sql
CREATE TABLE gdp_five_yrs AS
SELECT
    Country_name,
    MAX(CASE WHEN year = '2020' THEN gdp_per_capita_billion END) AS GDP_2020,
    MAX(CASE WHEN year = '2021' THEN gdp_per_capita_billion END) AS GDP_2021,
    MAX(CASE WHEN year = '2022' THEN gdp_per_capita_billion END) AS GDP_2022,
    MAX(CASE WHEN year = '2023' THEN gdp_per_capita_billion END) AS GDP_2023,
    MAX(CASE WHEN year = '2024' THEN gdp_per_capita_billion END) AS GDP_2024
FROM wb_africa
GROUP BY Country_name
```

```python
%%sql
SELECT * FROM gdp_five_yrs
WHERE country_name IN ('Kenya','South Africa','Nigeria','Egypt','Ethiopia')
```

```python
%%sql
SELECT
    Country_name,
    Gdp_2024,
    Gdp_2020,
    gdp_2024 - gdp_2020 AS growth,
    RANK() OVER(ORDER BY  gdp_2024 - gdp_2020  DESC)AS growth_RANK
FROM gdp_five_yrs
```

```python
%%sql
SELECT
    Country_name,
    Gdp_2024,
    Gdp_2023,
    gdp_2024 - gdp_2023 AS growth,
    RANK() OVER(ORDER BY  gdp_2024 - gdp_2023  DESC)AS growth_RANK
FROM gdp_five_yrs
```
