### Question 5

How many movies are there that are both in the "Thriller" genre and contain the word “love” anywhere in the keywords?


**Options:**
 - 48
 - 38
 - 14
 - 1

```python
%%sql
SELECT
 COUNT( DISTINCT m.movie_id)
FROM Movies m
LEFT JOIN genremap gm
    ON gm.movie_id = m.movie_id
LEFT JOIN genres g 
    ON g.genre_id = gm.genre_id
LEFT JOIN keywordmap km
    ON km.movie_id = m.movie_id
LEFT JOIN keywords k
     ON k.keyword_id = km.keyword_id
WHERE 
    g.genre_name = 'Thriller' 
AND
    k.keyword_name LIKE '%love%'
```

### Question 7

How many unique characters has "Vin Diesel" played so far in the database?

**Options:**
 - 24
 - 19
 - 18
 - 16

```python
%%sql
SELECT 
 COUNT(a.actor_name) AS no_of_occurence
FROM actors a
LEFT JOIN casts c
    ON c.actor_id = a.actor_id
LEFT JOIN movies m
    ON m.movie_id = c.movie_id
WHERE  actor_name = 'Vin Diesel'
```

### Question 8

What are the genres of the movie “The Royal Tenenbaums”?


**Options:**
 - Action, Romance
 - Drama, Comedy
 - Crime, Thriller
 - Drama, Romance

```python
%%sql
SELECT 
    genre_name
FROM genres g
JOIN genremap gm
    ON gm.genre_id = g.genre_id
JOIN movies m
    ON m.movie_id = gm.movie_id
WHERE title  = 'The Royal Tenenbaums'
```

### Question 9

What are the three production companies that have the highest movie popularity score on average, as recorded within the database?


**Options:**

 - MCL Films S.A., Turner Pictures, and George Stevens Productions
 - The Donners' Company, Bulletproof Cupid, and Kinberg Genre
 - Bulletproof Cupid, The Donners' Company, and MCL Films S.A
 - B.Sting Entertainment, Illumination Pictures, and Aztec Musique

```python
%%sql
SELECT
    AVG(m.popularity) AS popularity_avg,
    pc.production_company_name

FROM movies m

LEFT JOIN productioncompanymap pm
    ON pm.movie_id = m.movie_id
LEFT JOIN productioncompanies pc
    ON pc.production_company_id = pm.production_company_id
GROUP BY pc.production_company_name
ORDER BY popularity_avg DESC
limit 3;
    

    
```

### Question 10

### Question 11

Which genre has, on average, the lowest movie popularity score? 


**Options:**

 - Science Fiction
 - Animation
 - Documentary
 - Foreign

```python
%%sql
SELECT
    AVG(m.popularity) AS popularity_avg,
   g.genre_name
FROM Movies m
JOIN genremap gm
    ON gm.movie_id = m.movie_id
JOIN genres g
    ON g.genre_id = gm.genre_id
GROUP BY g.genre_id
ORDER BY popularity_avg ASC
```

### Question 12

Which award category has the highest number of actor nominations (actors can be male or female)? (Hint: `Oscars.name` contains both actors' names and film names.)

**Options:**

- Special Achievement Award
- Actor in a Supporting Role
- Actress in a Supporting Role
- Best Picture



```python
%%sql
SELECT
     COUNT(name) AS no_of_actors,
        award
FROM oscars
GROUP BY award
ORDER BY  no_of_actors desc
limit 5;
```
